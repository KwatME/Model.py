```python
w = pd.read_table(PATH_DICT["w_file_path"], index_col=0)

h = pd.read_table(PATH_DICT["h_file_path"], index_col=0)
```

```python
state_x_feature = ccal.make_membership_df_from_categorical_series(
    pd.read_table(PATH_DICT["w_hcc__k_x_column_file_path"], index_col=0).loc[
        "K{}".format(W_HCC_K)
    ]
)

state_x_feature.index = ("S{}".format(i) for i in state_x_feature.index)

state_x_sample = ccal.make_membership_df_from_categorical_series(
    pd.read_table(PATH_DICT["h_hcc__k_x_column_file_path"], index_col=0).loc[
        "K{}".format(H_HCC_K)
    ]
)

state_x_sample.index = ("S{}".format(i) for i in state_x_sample.index)
```

```python
feature_dicts = make_feature_dicts()
```

```python
for element in ("feature", "sample"):

    feature_dicts[element] = {
        key: value
        for key, value in feature_dicts[element].items()
        if value["data_type"] != "categorical"
    }

    print("{}:{}{}".format(element, "\n\t", "\n\t".join(feature_dicts[element].keys())))
```

```python
make_match_panels_keyword_arguments = dict(
    n_job=MAX_N_JOB,
    n_required_for_match_function=0.64,
    extreme_feature_threshold=EXTREME_FEATURE_THRESHOLD,
    plot_std=PLOT_STD,
)
```

```python
for target_x_feature, data_type, directory_path in (
    (w.T, "continuous", PATH_DICT["w_match_directory_path"]),
    (state_x_feature, "binary", PATH_DICT["w_hcc_match_directory_path"]),
):

    ccal.make_match_panels(
        target_x_feature,
        feature_dicts["feature"],
        target_type=data_type,
        directory_path=directory_path,
        plotly_directory_path=PATH_DICT["w_match_plotly_directory_path"],
        **make_match_panels_keyword_arguments,
    )
```

```python
for target_x_sample, data_type, directory_path in (
    (h, "continuous", PATH_DICT["h_match_directory_path"]),
    (state_x_sample, "binary", PATH_DICT["h_hcc_match_directory_path"]),
):

    ccal.make_match_panels(
        target_x_sample,
        feature_dicts["sample"],
        target_type=data_type,
        directory_path=directory_path,
        plotly_directory_path=PATH_DICT["h_match_plotly_directory_path"],
        **make_match_panels_keyword_arguments,
    )
```
